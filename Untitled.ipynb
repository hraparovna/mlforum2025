```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python

# ---------- 1.      ----------
np.random.seed(42)
n = 8000

mean1, cov1 = (-0.6, 0.8), [[0.07,  0.06], [0.06, 0.08]]
mean2, cov2 = ( 0.7, -0.3), [[0.05, -0.04], [-0.04, 0.05]]
mean3, cov3 = ( 0.0,  0.0), [[0.04,  0.00], [0.00, 0.04]]

s1 = np.random.multivariate_normal(mean1, cov1, n // 3)
s2 = np.random.multivariate_normal(mean2, cov2, n // 3)
s3 = np.random.multivariate_normal(mean3, cov3, n // 3 + n % 3)
samples = np.vstack((s1, s2, s3)).T            # shape (2, N)  gaussian_kde

# ---------- 2. KDE ----------
kde = gaussian_kde(samples, bw_method="scott")
xmin, xmax, ymin, ymax = -1.5, 1.5, -1.5, 1.5
xx, yy = np.mgrid[xmin:xmax:200j, ymin:ymax:200j]
zz = kde(np.vstack([xx.ravel(), yy.ravel()])).reshape(xx.shape)

# ---------- 3.  ----------
plt.figure(figsize=(8, 8))
plt.contourf(xx, yy, zz, levels=100, cmap="coolwarm", alpha=0.75)  # KDE heatmap

#  
pts = np.array([[-1, -1], [0, 0], [1, 0], [0, 1]])
plt.scatter(pts[:, 0], pts[:, 1], s=180, edgecolors="k", linewidths=1.6, zorder=5)

plt.xlim(xmin, xmax); plt.ylim(ymin, ymax)
plt.xlabel("X"); plt.ylabel("Y")
plt.title("Joint KDE with Highlighted Points")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
